## Отбор по списку признаков <code> isin </code>

Метод isin вычисляет булев вектор членства в множестве и может быть полезен для фильтрации набора данных относительно подмножества значений в объекте Series или столбце DataFrame:

```python
obj = pd.Series(['c', 'a', 'd', 'a', 'a', 'b', 'b', 'c', 'c'])
mask = obj.isin(['b', 'c'])
mask
```

```python
obj[mask]
```

Короткий вариант кода:

```python
obj[obj.isin(['b', 'c'])]
```

## Сводные таблицы

Агрегирует таблицу по одному или нескольким ключам и строит другую таблицу, в которой одни групповые ключи расположены в строках, а другие – в столбцах.

Более простой вариант работы с табличными данными, если на выходе нужна именно тыблица. 

```python
df = pd.read_excel("datasets/sales-funnel.xlsx")
df.head()
```

```python
cat_name = df["Status"].unique()
df["Status"] = df["Status"].astype("category")
df["Status"].cat.set_categories(cat_name,inplace=True)
df
```

```python
pd.pivot_table(df,index=["Name"])
```

```python
pd.pivot_table(df,index=["Name", "Rep", "Manager"]) # Rep - рекламный представитель
# итог не очень понятен
```

```python
# а так намного лучше
pd.pivot_table(df,index=["Manager", "Rep",  "Name"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
```

```python
# посчитаем накопленную сумму
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price","Quantity"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

```python
pd.pivot_table(df,index=["Manager","Status"],values=["Price"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":np.sum},fill_value=0)
```

```python
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
```
